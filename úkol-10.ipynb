## Zadání 1

Načti data o zaměstnancích z CSV souborů do tabulek (DataFrame). Ke každé tabulce přidej nový sloupec mesto, které bude obsahovat informaci o tom, ve kterém městě zaměstnanec pracuje.

```python
import pandas
praha = pandas.read_csv("/Users/petrakrykorkova/Documents/Programování/zam_praha.csv")
praha["mesto"] = "Praha"
praha
```

```python
plzen = pandas.read_csv("/Users/petrakrykorkova/Documents/Programování/zam_plzeň.csv")
plzen["mesto"] = "Plzeň"
plzen
```

```python
liberec = pandas.read_csv("/Users/petrakrykorkova/Documents/Programování/zam_liberec.csv")
liberec["mesto"] = "Liberec"
liberec
```

Vytvoř novou tabulku zamestnanci a ulož do ní informace o všech zaměstnancích (operace concat).

```python
zamestnanci = pandas.concat([praha, plzen, liberec])
zamestnanci = zamestnanci.reset_index(drop=True)
#zamestnanci.info()
zamestnanci
```

Ze souboru platy_2021_02.csv načti platy zaměstnanců za únor 2021. Propoj tabulku (operace join) s platy a tabulku se zaměstnanci pomocí sloupce cislo_zamestnance.

```python
mzdy_unor = pandas.read_csv("/Users/petrakrykorkova/Documents/Programování/platy_2021_02.csv")
mzdy_unor
```

```python
zamestnanci_mzdy = pandas.merge(zamestnanci, mzdy_unor, on="cislo_zamestnance", how="left")
zamestnanci_mzdy
```

Porovnej rozměry tabulek před spojením a po spojení. Pokud nemá některý zaměstnanec plat za únor, znamená to, že v naší firmě již nepracuje.

```python
rozmer_z = zamestnanci.shape[0] == zamestnanci_mzdy.shape[0]
rozmer_z
```

```python
rozmer_unor = mzdy_unor.shape[0]
rozmer_unor
```

Spočti průměrný plat zaměstnanců v jednotlivých kancelářích.

```python
prumer_mesta = zamestnanci_mzdy.groupby("mesto")["plat"].mean()
prumer_mesta
```

```python
zakaznik = pandas.read_csv("/Users/petrakrykorkova/Documents/Programování/vykazy.csv")
zakaznik
```
